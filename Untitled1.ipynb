```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r"C:\Users\lenovo\OneDrive\Desktop\ML PROJECTS csvs\Mall_Customers.csv")
df.head()
```

```python
#checking our dataset
df.info()
df.describe()
```

```python
#visualizing the dataset 
sns.pairplot(df, diag_kind='kde')
plt.suptitle("Feature Relationships", y=1.02)
plt.show()
```

```python
#Clustering selection 
X = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Scaling to make K- means work better Using standardscaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


```

```python
# finding the number of clusters using elbow method 
inertia = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
#Appling K-means in clusters K=5
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
df.head()
```

```python
import os
os.environ["OMP_NUM_THREADS"] = "1"
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='Annual Income (k$)', y='Spending Score (1-100)',
    hue='Cluster', palette='tab10', data=df, s=100
)
plt.title("Customer Segments by Income and Spending Score")
plt.show()
```

```python
#Analysing each cluster
#Cluster 0  High income, high spenders
#Cluster 1  Low income, low spenders,
#Cluster 2  Young, moderate spenders 
df.groupby('Cluster').mean()
```
